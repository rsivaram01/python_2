```python
color = 'aaa'

if color == 'red':
    print("the color is red")
elif color == 'orange':
    print("the color is orange")
elif color == 'white':
    print("the color is white")
elif color == 'blue':
    print("the color is blue")
else:
    print("im not sure about the color")
```
