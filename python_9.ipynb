```python
# Inheritance:-
=============

Creating a new class from an existing class is called inheritance

# 1. Super class, Sub class
# 2. Parent Class, Child Class

```

```python
class parentclass:
    var1 = "im var1"
    var2 = 'im var2'

class childclass(parentclass):
    var3 = 'im var3'
    
```

```python
pobj = parentclass()

cobj = childclass()
```

```python
pobj.var1
```

```python
pobj.var2
```

```python
pobj.var3
```

```python
cobj.var3
```

```python
cobj.var1
```

```python
cobj.var2
```

```python
class dad:
    a = 'im dad'

class mom:
    b = 'im mom'

class child(dad,mom):
    c = 'im child'
```

```python
childobj = child()
```

```python
childobj.a
```

```python
class dog:
    """ A sample class to demo dog """
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def intro(self):
        print(f"my dog name is {self.name}")
        print(f"my dog age is {self.age}")

    def sit(self):
        print(f"{self.name} is now sitting")

    def rollover(self):
        print(f"{self.name} is rollingover now")

```

```python
ashok_dog = dog('bruno', 5)
```

```python
ashok_dog.sit()
```

```python
ashok_dog.intro()
```

```python
naveens_dog = dog('chotu', 7)
```

```python
naveens_dog.intro()
```

```python
# file handling in python:-
=========================

1. how to read a file
2. how to write something to a file

```

```python
with open("sample.txt") as file_obj:
    contents = file_obj.read()

print(contents)
    
```

```python
# reading a file line by line

# for loop

with open("sample.txt",'r') as file_obj:
    for i in file_obj:
        print(i)

```

```python
# wirting data to a file:-
# ======================

with open("sample.txt",'w') as file_obj:
    read = file_obj.write("I love python")


```

```python
# append data to a file:-
# ======================

with open("sample.txt",'a') as file_obj:
    read = file_obj.write("\nthis is for testing")


```
