```python
What is Sqoop?

Sqoop is an import and export tool, which is used to import data from jdbc supported databases to hadoop or vice verse 


my sql or oracle or sql server ===============> hadoop   -----> import 

Default format of importing data in sqoop is - csv

hadoop  ======================================> my sql or oracle or sql server   -----> export
```

```python
# req:- importing data with tab delimeter ====> \t


sqoop import --connect  jdbc:mysql://hostname --username user1 --password passcode --table sales.belt -m 3 --target-dir /Data/sales/belt/ \
--fields-terminated-by '\t' --enclosed-by '"'

```

```python
# req:- I don't want to import all columns , i want to import specfic columns =====> I want 3 columns (id,name,year)

sqoop import --connect  jdbc:mysql://hostname --username user1 --password passcode --table sales.belt --columns "id,name,year" -m 3 --target-dir /Data/sales/belt/ \
--fields-terminated-by '\t' --enclosed-by '"'
```

```python
# req:- I don't want to import all columns , i want to import specfic columns with filter =====> I want 3 columns (id,name,year)

sqoop import --connect  jdbc:mysql://hostname --username user1 --password passcode --table sales.belt --columns "id,name,year" --where "year>2019" \
-m 3 --target-dir /Data/sales/belt/ \
--fields-terminated-by '\t' --enclosed-by '"'
```

```python
# sql ---> select * from email where len(email) > 30

sqoop import --connect  jdbc:mysql://hostname --username user1 --password passcode --query "select * from email where len(email)>30"  \
-m 3 --target-dir /Data/sales/belt/ \
--fields-terminated-by '\t' --enclosed-by '"'
```
