```python
Tables :-
========

1. Managed Table or Internal Table
2. External Table
```

```python
CREATE TABLE employee_orc (
    emp_id        INT,
    emp_name      STRING,
    department    STRING,
    salary        DOUBLE,
    joining_date  DATE
)
STORED AS ORC;
```

```python
CREATE TABLE employee_csv (
    emp_id        INT,
    emp_name      STRING,
    department    STRING,
    salary        DOUBLE,
    joining_date  DATE
)
ROW FORMAT DELIMITED
FIELDS TERMINATED BY ','
STORED AS TEXTFILE;
```

```python
# External table


CREATE EXTERNAL TABLE employee_orc_ext (
    emp_id        INT,
    emp_name      STRING,
    department    STRING,
    salary        DOUBLE,
    joining_date  DATE
)
STORED AS ORC
LOCATION '/data/hive/employee_orc/';
```

```python
select * from employee_orc_ext
```

```python
Internal Tables:-
================
1. Metadata and table data will be managed by hive
2. when we drop internal or managed table schema and data of the table will be deleted

drop table employee_orc_inter


External Tables:-
================
1. Only metadata will be managed by hive
2. when we drop external table only schema will get dropped, data won't be deleted

drop table employee_orc_exter
```
